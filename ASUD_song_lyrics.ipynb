```python
#Subset Stratifying
group_Pop_sizes = dfEnPop10['year'].value_counts(normalize=True)
total_samples = 100000
Pop_samples_per_group = (group_Pop_sizes * total_samples).round().astype(int)
subsetPop_df = pd.DataFrame() 

for group, size in Pop_samples_per_group.items():
    group_Pop_sample = dfEn10[dfEn10['year'] == group].sample(n=size, replace=True)
    subsetPop_df = pd.concat([subsetPop_df, group_Pop_sample])

subsetPop_df.head()
```
